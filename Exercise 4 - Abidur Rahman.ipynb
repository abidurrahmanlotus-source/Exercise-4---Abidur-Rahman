### 2. Function to calculate the area of a circle

```python
import math

#Calculate Circle Area with Radius Input
def circle_area(radius):
    return math.pi * radius ** 2

print(circle_area(10))
```

### 3. Simple Calculator Function

```python
#Calculator Function for Two Numeric Inputs and Operator Input
def calculator(num1, num2, operation):
    if operation == 'a':
        return num1 + num2
    elif operation == 's':
        return num1 - num2
    elif operation == 'm':
        return num1 * num2
    elif operation == 'd':
        return num1 / num2
    else:
        return "Invalid Input."

print(calculator(2, 5, 'd'))
```

### 4. Rectangle Class

```python
#Class for Rectangle
class Rectangle:
    #Set Length and Width
    def __init__(self, length, width):
        self.length = length
        self.width = width

    #Calculate Area
    def area(self):
        return self.length * self.width

# Input and Output
r = Rectangle(5, 10)
print(r.area())
```

### 5. Shape and Square Classes

```python
#Class
class Shape:
    #Set Length and Width
    def __init__(self, name, length):
        self.name = name
        self.length = length

    #Calculate Area
    def area(self):
        return 0

#Sub Class
class Square(Shape):
    #Set Length and Width
    def __init__(self, name, length):
        super().__init__(name, length)

    #Calculate Area
    def area(self):
        print("The area is:")
        return self.length * self.length

    #Describes Square
    def describe(self):
        return f"This is a: {self.name}"


# Input and Output
s = Square('square', 5)
print(s.area())
print(s.describe())
```
